```python
import nltk
from nltk.corpus import stopwords
```

#### preprocessing stuff

```python

generated_tokens = []
for i in range(57280) :
    string = df.Cleandata[i] + df.Summary[i]
    tokens = nltk.tokenize.word_tokenize(string)
    lower_tokens = [token.lower() for token in tokens]

    stop = stopwords.words('english')
    tokens1 = [token for token in lower_tokens if token not in stop]

    lmtzr = nltk.stem.WordNetLemmatizer()
    tokens2 = [lmtzr.lemmatize(token) for token in tokens1]
        
    tokens3 = " ".join(str(x) for x in tokens2)
    generated_tokens.append(tokens3)
```

```python
df['tokens'] = generated_tokens
```
