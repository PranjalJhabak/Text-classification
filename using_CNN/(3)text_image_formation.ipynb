```python
string = df['Cleandata'][0]

string = re.sub("[^\w]", " ",  string).split()
lower_tokens = [token.lower() for token in string]
    
lmtzr = nltk.stem.WordNetLemmatizer()
tokens = [lmtzr.lemmatize(token) for token in lower_tokens]
    
tok_corp= [nltk.word_tokenize(sent.decode('utf-8')) for sent in tokens]

p = model.wv[tok_corp[0]]
for i in range(1,len(tok_corp)-1):    
        s = model.wv[tok_corp[i]]
        p = np.append(p,s,axis =0)
n = len(p[:,1])    
a = np.zeros(shape=(200-n,10))
t = np.append(p,a,axis =0)
for j in range(1,57279):
    print j
    string = df['Cleandata'][j]
    string = re.sub("[^\w]", " ",  string).split()
    lower_tokens = [token.lower() for token in string]
    
    lmtzr = nltk.stem.WordNetLemmatizer()
    tokens = [lmtzr.lemmatize(token) for token in lower_tokens]
    
    tok_corp= [nltk.word_tokenize(sent.decode('utf-8')) for sent in tokens]
     
    if tok_corp[0][0] in model: 
        q = model.wv[tok_corp[0]]
    else :
        q = [np.zeros(10)]
    
    m = 0
    for k in range(0,len(tok_corp)):
        m = m + len(tok_corp[k])
    
                
                
    if m <= 200 :

        for i in range(1,len(tok_corp)):   
            if tok_corp[i][0] in model:
                s = model.wv[tok_corp[i]]
            else :
                s = [np.zeros(10)]
                
            q = np.append(q,s,axis =0)
        n = len(q[:,1])
        a = np.zeros(shape=(200-n,10))
        q = np.append(q,a,axis = 0)
        
    else :
        
        m = 0
        for k in range(0,200):
            m = m + len(tok_corp[k])
            if m >= 200 :
                break 
              
        for i in range(1,k+1): 
            if tok_corp[i][0] in model:
                s = model.wv[tok_corp[i]]
            else :
                s = [np.zeros(10)]
            q = np.append(q,s,axis =0) 
    t = np.dstack((t, q))
t.shape    
```
